```python
# Define the Mathematical Fuzzy Intensification Operator
def apply_fuzzy_logic(img):
    img_norm = img.astype(np.float32) / 255.0
    enhanced = np.where(img_norm <= 0.5, 2 * (img_norm ** 2), 1 - 2 * ((1 - img_norm) ** 2))
    return (enhanced * 255).astype(np.uint8)

print("=> Applying Fuzzy Logic to Dataset. This will take a few minutes...")

# Apply fuzzy logic to both training and validation sets
for folder in [data['train'], data['val']]:
    for img_path in glob.glob(os.path.join(folder, '*.*')):
        # Skip hidden Mac OS files if they snuck through
        if "__MACOSX" in img_path:
            continue

        img = cv2.imread(img_path)
        if img is not None:
            cv2.imwrite(img_path, apply_fuzzy_logic(img))

print("=> FUZZY PREPROCESSING COMPLETE. Images are ready for the neural network.")
```

```python
import os
import glob
import yaml

# Exact path from your Colab logs
yaml_path = '/content/dataset/sc_project /data.yaml'

# 1. Load the original YAML to extract exact class IDs
with open(yaml_path, 'r', encoding='utf-8') as f:
    data = yaml.safe_load(f)

old_names = data.get('names', [])

# 2. Define 5 new robust, high-frequency categories
new_categories = ['Hopper_Cicada', 'Aphid', 'Borer', 'Worm_Caterpillar', 'Beetle_Weevil']

# 3. Dynamically map old integer IDs to new category IDs
class_mapping = {}
for old_id, name in enumerate(old_names):
    name_lower = name.lower()
    if any(kw in name_lower for kw in ['hopper', 'cicada', 'lygus']):
        class_mapping[old_id] = 0
    elif 'aphid' in name_lower:
        class_mapping[old_id] = 1
    elif 'borer' in name_lower:
        class_mapping[old_id] = 2
    elif any(kw in name_lower for kw in ['worm', 'grub', 'spodoptera', 'moth', 'cutworm', 'armyworm']):
        class_mapping[old_id] = 3
    elif any(kw in name_lower for kw in ['beetle', 'weevil', 'cantharis', 'chafer', 'flea']):
        class_mapping[old_id] = 4

print(f"[*] Mapped {len(class_mapping)} sparse classes into 5 consolidated categories.")

# 4. Function to rewrite the .txt labels and drop unmapped/rare classes
def consolidate_labels(label_dir):
    txt_files = glob.glob(os.path.join(label_dir, '*.txt'))
    modified_count = 0

    for txt_path in txt_files:
        with open(txt_path, 'r') as file:
            lines = file.readlines()

        new_lines = []
        for line in lines:
            parts = line.strip().split()
            if not parts: continue

            old_id = int(parts[0])
            if old_id in class_mapping:
                new_id = class_mapping[old_id]
                new_lines.append(f"{new_id} {' '.join(parts[1:])}\n")

        # Overwrite file with new mapped IDs
        with open(txt_path, 'w') as file:
            file.writelines(new_lines)
        modified_count += 1

    return modified_count

# 5. Execute mapping
train_dir = '/content/dataset/sc_project /train/labels'
val_dir = '/content/dataset/sc_project /valid/labels'

if os.path.exists(train_dir) and os.path.exists(val_dir):
    print(f"[*] Cleaned {consolidate_labels(train_dir)} training labels.")
    print(f"[*] Cleaned {consolidate_labels(val_dir)} validation labels.")

    # 6. Generate the new balanced YAML file
    data['nc'] = len(new_categories)
    data['names'] = new_categories
    new_yaml_path = '/content/dataset/sc_project /data_balanced.yaml'

    with open(new_yaml_path, 'w', encoding='utf-8') as f:
        yaml.dump(data, f, allow_unicode=True)

    print("\n" + "="*50)
    print(f" ✅ SUCCESS. Your dataset is fixed.")
    print(f" Start your new training run using:")
    print(f" model.train(data='{new_yaml_path}', epochs=100, imgsz=640, patience=20)")
    print("="*50)
else:
    print("\n[!] ERROR: Could not find label directories. Check the folder paths.")
```

```python
import os
import cv2
import glob
import random
import matplotlib.pyplot as plt
from ultralytics import YOLO

# 1. THE CORE TESTING FUNCTION
def test_any_image(custom_image_path, model_path):
    print("\n" + "="*50)
    print(f" ‑‑ INFERENCE: {os.path.basename(custom_image_path)}")
    print("="*50)

    if not os.path.exists(custom_image_path):
        print(f" [!] ERROR: File not found at {custom_image_path}")
        return

    # Load the specific model weights
    eval_model = YOLO(model_path)

    # Read the raw image (NO FUZZY LOGIC)
    raw_img = cv2.imread(custom_image_path)
    if raw_img is None:
        print(" [!] ERROR: Unable to read image file.")
        return

    print(" [*] Running Neural Network Engine on RAW image...")
    # Lowered confidence threshold to 0.15 to expose weak guesses
    results = eval_model.predict(source=raw_img, conf=0.15, verbose=False)

    boxes = results[0].boxes
    if len(boxes) == 0:
        print("\n [✓] RESULT: NO PEST DETECTED.")
    else:
        print(f"\n [!] RESULT: PEST DETECTED! ({len(boxes)} instances found)")
        for box in boxes:
            cls_id = int(box.cls[0])
            conf = float(box.conf[0])
            print(f"     -> Species: {eval_model.names[cls_id]} | Confidence: {conf*100:.2f}%")

    # Plot and display
    final_img_rgb = cv2.cvtColor(results[0].plot(), cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(6,6))
    plt.imshow(final_img_rgb)
    plt.title(f"F-YOLO Output (Raw Image)")
    plt.axis('off')
    plt.show()

# 2. SET CORRECT PATHS
# Pointing to the correct path where training results were saved
LATEST_MODEL = '/content/runs/fyolo_fast_train/weights/best.pt'
VALIDATION_DIR = '/content/dataset/sc_project /valid/images'

# ---------------------------------------------------------
# TEST 1: AUTOMATIC 5 RANDOM VALIDATION IMAGES
# ---------------------------------------------------------
print("\n>>> STARTING TEST 1: 5 RANDOM VALIDATION IMAGES")
val_images = glob.glob(os.path.join(VALIDATION_DIR, '*.jpg'))
# Filter out Mac OS hidden files
val_images = [img for img in val_images if "__MACOSX" not in img]

if val_images:
    for i in range(5):
        print(f"\n" + "*"*50)
        print(f" TESTING RANDOM IMAGE {i+1} OF 5")
        print("*"*50)
        random_test_img = random.choice(val_images)
        test_any_image(random_test_img, LATEST_MODEL)
else:
    print(f" [!] Error: Could not find validation images in {VALIDATION_DIR}")

# ---------------------------------------------------------
# TEST 2: MANUAL CUSTOM IMAGE
# ---------------------------------------------------------
print("\n>>> STARTING TEST 2: MANUAL CUSTOM IMAGE")
my_custom_image_path = '/content/dataset/sc_project /valid/images/IP001000001.jpg'

test_any_image(my_custom_image_path, LATEST_MODEL)
```
